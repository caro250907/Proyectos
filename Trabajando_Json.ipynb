En los valores se tiene un diccionario de datos, esté valor se vuelve un dataframe.

```
df_final=pd.DataFrame(df['Afghanistan'][0],columns=['confirmed','date','deaths','recovered'],index=[0])
df_final['Pais']= df.columns[0]
```

```
columnas=df.columns
```

```

for c in columnas:
  for i in range(len(df)):
    df_aux=pd.DataFrame(df.loc[i,c],columns=['confirmed','date','deaths','recovered'],index=[0])
    df_aux['Pais']= c
    df_final=df_final.append(df_aux)
    

```
